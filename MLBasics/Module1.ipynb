```python
#reporting the no.of words in first column
from csv import reader
strings = []
with open('Data/Assignment_1_a/Reviews/cv000_29590.csv', 'r') as read_obj:
    csv_reader = reader(read_obj)
    header = next(csv_reader)
    if header != None:
        strings.append(header[0])
        for column in csv_reader:
            strings.append(column[0])
words = 0
for item in strings:
    words += (len(item.split()))
print('No.of words in first column is',words)
```

```python
#counting unique words in first column igonoring the punctuations
string = ' '.join(strings)
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
no_punct = ""
for char in string:
    if char not in punctuations:
        no_punct = no_punct + char
unique_words = list(set(no_punct.split()))
print('No.of unique words in first column ignoring the punctuations are',len(unique_words))
```
